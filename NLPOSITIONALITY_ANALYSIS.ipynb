```python
import pandas as pd
df = pd.read_csv('./data/nlpositionality_{}_{}.csv'.format("toxicity", "processed"))
df
```

```python
subset_df = df[df['action'].str.contains('(?i)fag|f\*g')]
subset_df
```

```python
annots = {}
for index, row in subset_df.iterrows():
    if row['action'] not in annots:
        annots[row['action']] = []
    annots[row['action']].append(row['litw'])
```

```python
for k, v in annots.items():
    print(k, len(v))
```

```python
uniq_annotators = subset_df['session_id'].unique()
print(uniq_annotators)
uniq_comments = subset_df['action'].unique()
print(uniq_comments)
```

```python
id_to_idx = {k : v for (k, v) in zip(uniq_annotators, range(len(uniq_annotators)))}
comment_to_idx = {k : v for (k, v) in zip(uniq_comments, range(len(uniq_comments)))}
```

```python
import torch

X = torch.zeros((len(uniq_annotators), len(uniq_comments)))
for _, row in subset_df.iterrows():
    X[id_to_idx[row['session_id']], comment_to_idx[row['action']]] = row['litw']
```

```python
def print_ratings(X, Z):

    for j in range(X.size(1)):
        print('Comment {}: {}'.format(j + 1, uniq_comments[j]))
        print('n = {}'.format(len(X[:, j][X[:, j] != 0])))

        print('Pre-interpolation mean:', torch.mean(X[:, j][X[:, j] != 0]).item())
        print('Pre-interpolation stdev:', torch.std(X[:, j][X[:, j] != 0]).item())

        print(X[:, j])
        print()

        print('Post-interpolation mean:', torch.mean(Z[:, j][Z[:, j] != 0]).item())
        print('Post-interpolation stdev:', torch.std(Z[:, j][Z[:, j] != 0]).item())

        print(Z[:, j])

        print()

# print_ratings(X, X)
```

```python
def interpolate_ratings(X, attrs = ['gender']):
    from torch_geometric.utils import coalesce, to_networkx
    from torch_geometric.transforms import FeaturePropagation
    from torch_geometric.data import Data
    import networkx as nx
    import matplotlib.pyplot as plt
    from grave import plot_network

    annotators_with_attr = {}

    for attr in attrs:

        if attr not in annotators_with_attr:
            annotators_with_attr[attr] = {}

        for _, row in subset_df.iterrows():
            if row[attr] == 'None':
                continue

            if row[attr] not in annotators_with_attr[attr]:
                annotators_with_attr[attr][row[attr]] = set()
            annotators_with_attr[attr][row[attr]].add(id_to_idx[row['session_id']])

    edge_index_dict = {}
    for attr in annotators_with_attr:
        for _, annotators in annotators_with_attr[attr].items():

            for a_i in annotators:
                for a_j in annotators:

                    if (a_i, a_j) in edge_index_dict:
                        edge_index_dict[(a_i, a_j)] += 1
                    else:
                        edge_index_dict[(a_i, a_j)] = 1

    edge_index = []
    edge_weight = []
    for (a_i, a_j), w in edge_index_dict.items():
        edge_index.append([a_i, a_j])
        edge_weight.append(w)

    edge_index, edge_weight = coalesce(torch.tensor(edge_index).long().t(), torch.tensor(edge_weight).float())
    print(edge_index, edge_weight)

    data = Data(x=X, edge_index=edge_index, edge_weight=edge_weight)

#     G = to_networkx(data, to_undirected=True)
#     plt.show()

    return FeaturePropagation(X == 0)(data).x
```

```python
attrs = ['age',
        'gender',
        'ethnicity',
        'religion',
        'education',
        'country_longest',
        'country_residence',
        'native_language']

Z = interpolate_ratings(X, attrs)
print_ratings(X, Z)
```
